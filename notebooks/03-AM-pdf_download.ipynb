```python
import requests

def clean(url):
    return url.replace('https://', '').replace('http://', '').replace('/', '_')
    
for url in df[~df.ee.isna()].ee.unique():
    if '.pdf' in url: # direct PDF link (e.g., CEUR)
        r = requests.get(url)
        print(url)
        with open('../data/raw/pdfs/' + clean(url), 'wb') as f:
            f.write(r.content)
    else:
        r = requests.get(url)
        for h in r.history:
            if '/chapter/' in h.url: #springer URL
                pdf_url = h.url.replace('chapter', 'content/pdf') + '.pdf'
                print(pdf_url)
                with open('../data/raw/pdfs/' + clean(pdf_url), 'wb') as f:
                    f.write(requests.get(pdf_url).content)
```

```python
def clean(url):
    return url.replace('https://', '').replace('http://', '').replace('/', '_')

#clean CEUR.WS
for index, row in df[~df.ee.isna()].iterrows():
    if '.pdf' in row['ee']: # direct PDF link (e.g., CEUR)
        old_path = '../data/raw/pdfs/' + clean(row['ee'])
        if os.path.exists(old_path):
            new_path = '../data/raw/pdfs/' + row['key'].replace('/', '_') + '.pdf'
            print(True, old_path, new_path)
            os.rename(old_path, new_path)
    else:
        r = requests.get(row['ee'])
        for h in r.history:
            if '/chapter/' in h.url: #springer URL
                old_path = '../data/raw/pdfs/' + clean(h.url.replace('chapter', 'content/pdf') + '.pdf')
                print(old_path)
                if os.path.exists(old_path):
                    new_path = '../data/raw/pdfs/' + row['key'].replace('/', '_') + '.pdf'
                    print(True, old_path, new_path)
                    os.rename(old_path, new_path)
```

```python
import requests

def clean(url):
    return url.replace('https://', '').replace('http://', '').replace('/', '_')
    
for url in df[~df.ee.isna()].ee.unique():
    if '.pdf' in url: # direct PDF link (e.g., CEUR)
        r = requests.get(url)
        print(url)
        with open('../data/raw/pdfs/' + clean(url), 'wb') as f:
            f.write(r.content)
    else:
        r = requests.get(url)
        for h in r.history:
            if '/chapter/' in h.url: #springer URL
                pdf_url = h.url.replace('chapter', 'content/pdf') + '.pdf'
                print(pdf_url)
                with open('../data/raw/pdfs/' + clean(pdf_url), 'wb') as f:
                    f.write(requests.get(pdf_url).content)
```

```python
from pathlib import Path
from grobid.models.form import Form, File
from grobid.models.response import Response
from grobid.client import Client, GrobidClientError

pdf_file = Path('../data/processed/pdf_1.pdf')
with open(pdf_file, "rb") as file:
    form = Form(
        file=File(
            payload=file.read(),
            file_name=pdf_file.name,
            mime_type="application/pdf",
        )
    )
    c = Client(base_url="https://kermitt2-grobid.hf.space", form=form)
    try:
        xml_content = c.sync_request().content  # TEI XML file in bytes
        print(xml_content)
    except GrobidClientError as e:
        print(e)
```
